```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
```

```python
s,p=1.0, .5
nbp=1000
error=np.arange(float(nbp))
lim=np.arange(float(nbp))
epsilon=0.25
n=100
for n in range(nbp):
    binomiale=np.random.binomial(s,p,n+1)
    #moyenne
    somme=0
    for i in binomiale:
        somme=somme+i
    moyenne=somme/float(n+1)
    #print(moyenne)
    error[n]=np.fabs(moyenne-p)


    #limite Hoeffding
    lim[n]=2*np.exp(-2*(n)*epsilon*epsilon)

```

```python
plt.plot(np.arange(nbp),lim,color='red')
plt.plot(np.arange(nbp),error)
plt.ylim(0.0,1.0)
plt.xlim(0.0,nbp)
plt.show()
```
